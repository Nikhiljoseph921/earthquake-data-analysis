```python
URL = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"
```

```python
df = pd.read_csv(URL)
df
```

```python
df.dropna(axis=0, how='all', inplace=True)
df
```

Keep necessary columns and  Select only the  columns from the DataFrame

```python

columns_to_keep = ['place', 'latitude', 'longitude', 'net', 'depthError', 'depth', 'type']
df_cleaned = df[columns_to_keep]
print(df_cleaned)
```

Define the function to extract the "territory' from 'place'

```python
def extract_country(place):
    if ',' in place:
        return place.split(',')[-1].strip() 
    else:
        return 'Unknown' 
df_cleaned['Territory'] = df_cleaned['place'].apply(extract_country)
print(df_cleaned[['place', 'Territory']])

```

Check for missing values in each column

```python
df_cleaned.isnull().sum()
```

fill some missing values occur in deptherror using median method

```python
median_depth_error = df_cleaned['depthError'].median()
df_cleaned['depthError'].fillna(median_depth_error, inplace=True)
print(f"Filled missing depthError values with the median: {median_depth_error}")
df_cleaned
```

 Check for missing values in each column

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned.info()
```

Covert them to a datatype

```python
df_cleaned["place"] = df_cleaned["place"].astype("string")
df_cleaned["Territory"] = df_cleaned["Territory"].astype("string")
df_cleaned["type"] = df_cleaned["type"].astype("string")
df_cleaned["net"] = df_cleaned["net"].astype("string")
```
